```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=df.rename(columns={'iyear':'year', 'imonth':'month', 'iday':'day', 'country_txt':'Country','region_txt':'Region','provstate':'State', 'attacktype1_txt':'attack', 'targtype1_txt':'targettype','targsubtype1_txt':'target', 'gname':'gang', 'weaptype1_txt':'weapon'})
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.corr()
```

```python
df['State']=df['State'].fillna('unknown')
df['target']=df['target'].fillna('unknown')
df
```

```python
plt.figure(figsize=(15,6))
plt.title('Attack type')
sns.histplot(x=df['attacktype1'])
```

```python
plt.figure(figsize=(15,6))
plt.title('Attacks month')
sns.scatterplot(x=df['month'],y=df['attacktype1'],marker="^",color='red')
```

```python
plt.figure(figsize=(15,6))
plt.title('Attacks days')
sns.scatterplot(x=df['day'],y=df['attacktype1'],color='green')
plt.show()
```

```python
plt.figure(figsize=(15,6))

plt.title('weapon Types')
sns.countplot(x=df['weapon'])
```

```python
pd.crosstab(df.year, df.Region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('No of Attacks')
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.title('weapon Types')
sns.countplot(x=df['weapon'])
```
